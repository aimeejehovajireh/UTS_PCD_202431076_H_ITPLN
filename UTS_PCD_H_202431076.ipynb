## IMPORT LIBRARY

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
gambar = cv.imread("tulisan.jpeg")
```

## DETEKSI WARNA (SOAL 1)

### CITRA KONTRAS

```python
kontras = cv.cvtColor(gambar, cv.COLOR_BGR2RGB)
fig, axs = plt.subplots(3, figsize=(8,10))
axs[0].imshow(kontras)
axs[1].hist(kontras.ravel(),256,[0,256])
hist = cv.calcHist([kontras],[0],None,[256],[0,256])
axs[2].plot(hist)

plt.show()
```

## MENCARI NILAI AMBANG (SOAL 2)

```python
gambar
```

```python
gambar.shape
print(gambar.dtype)
```

```python
b = gambar[:,:,0]
g = gambar[:,:,1]
r = gambar[:,:,2]
```

```python
panjang = len(gambar)
lebar = len(gambar[0])
 
citra_gray = np.zeros((panjang, lebar))
 
for i in range (panjang):
    for j in range (lebar):
        citra_gray[i,j] = 0.3989 * r[i,j] + 0.587 * g[i,j] + 0.1141 * b[i,j]
citra_gray = citra_gray.astype(np.uint8)
```

```python
panjang = len(gambar)
lebar = len(gambar[0])
ambang = 255
gambar_citra = np.zeros((panjang,lebar))
for i in range(panjang):
    for j in range(lebar):
        if citra_gray[i,j] > ambang:
            gambar_citra[i,j] = 1
        else:
            gambar_citra[i,j] = 0
plt.imshow(gambar_citra, cmap='gray')
```

### NILAI AMBANG BIRU

```python
ambang_bawah_biru = 95
ambang_atas_biru = 255
gambar_b = np.zeros((panjang, lebar), dtype=np.uint8)
for i in range(panjang):
    for j in range(lebar):
        if ambang_bawah_biru <= b[i, j] <= ambang_atas_biru and g[i, j] < 100 and r[i, j] < 100:
            gambar_b[i, j] = 1
        else:
            gambar_b[i, j] = 0
plt.imshow(gambar_b, cmap='gray')
```

### NILAI AMBANG BIRU & MERAH

```python
ambang_bawah_merah = 120
ambang_atas_merah = 255
gambar_m = np.zeros((panjang, lebar), dtype=np.uint8)
for i in range(panjang):
    for j in range(lebar):
        biru = ambang_bawah_biru <= b[i, j] <= ambang_atas_biru and r[i, j] < 100 and g[i, j] < 100
        merah = ambang_bawah_merah <= r[i, j] <= ambang_atas_merah and g[i, j] < 100 and b[i, j] < 100
        if merah or biru:
            gambar_m[i, j] = 1
        else:
            gambar_m[i, j] = 0 

plt.imshow(gambar_m, cmap='gray')
```

### NILAI AMBANG BIRU, HIJAU & MERAH

```python
ambang_bawah_hijau = 95
ambang_atas_hijau = 255
gambar_bgr = np.zeros((panjang, lebar), dtype=np.uint8)
for i in range(panjang):
    for j in range(lebar):
        biru = ambang_bawah_biru <= b[i, j] <= ambang_atas_biru and r[i, j] < 100 and g[i, j] < 100
        hijau = ambang_bawah_hijau <= g[i, j] <= ambang_atas_hijau and r[i, j] < 100 and b[i, j] < 100
        merah = ambang_bawah_merah <= r[i, j] <= ambang_atas_merah and g[i, j] < 100 and b[i, j] < 100
        if merah or biru or hijau:
            gambar_bgr[i, j] = 1
        else:
            gambar_bgr[i, j] = 0 

plt.imshow(gambar_bgr, cmap='gray')
```

```python
fig, axs = plt.subplots(2,2, figsize=(10, 8))

axs[0,0].imshow(gambar_citra, cmap='gray')
axs[0,0].set_title("Ambang None")
axs[0,0].axis('off')

axs[0,1].set_title("Ambang Biru")
axs[0,1].imshow(gambar_b, cmap='gray')

axs[1,0].set_title("Ambang Biru dan Merah")
axs[1,0].imshow(gambar_m, cmap='gray')

axs[1,1].set_title("Ambang Biru, Hijau, dan Merah")
axs[1,1].imshow(gambar_bgr, cmap='gray')

plt.tight_layout()
plt.show()
```

## ANALISIS KERNEL KONVOLUSI (SOAL 3)

202431076_AIMEEJEHOVAJIREH (KERNEL 1 = 2 & KERNEL 2 = 5)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('selfie.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
kernel1 = np.array([[-2, -1,  0],
                    [-1,  1,  1],
                    [ 0,  1,  2]])
```

```python
kernel2 = np.array([[-1,  0,  1],
                    [-2,  0,  2],
                    [-1,  0,  1]])
```

```python
res_kernel1 = cv2.filter2D(img_gray, -1, kernel1)
res_kernel2 = cv2.filter2D(img_gray, -1, kernel2)
```

```python
plt.figure(figsize=(15, 5))
```

### Menampilkan Hasil Citra

```python
plt.subplot(1, 3, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Citra Asli (Grayscale)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(res_kernel1, cmap='gray')
plt.title('Hasil Kernel 1 (No. 2)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(res_kernel2, cmap='gray')
plt.title('Hasil Kernel 2 (No. 5)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

### Menampilkan Histogram untuk Analisis Intensitas 

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(img_gray.ravel(), 256, [0, 256])
plt.title('Histogram Asli')

plt.subplot(1, 3, 2)
plt.hist(res_kernel1.ravel(), 256, [0, 256])
plt.title('Histogram Kernel 1')

plt.subplot(1, 3, 3)
plt.hist(res_kernel2.ravel(), 256, [0, 256])
plt.title('Histogram Kernel 2')

plt.tight_layout()
plt.show()
```
